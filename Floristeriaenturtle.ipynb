```python
import turtle

# Configuración inicial
screen = turtle.Screen()
screen.title("Floristería con jardín interior por zonas")
screen.bgcolor("lightyellow")
t = turtle.Turtle()
t.speed(0)

#  Funciones básicas
def dibujar_rectangulo(t, ancho, alto, color):
    t.fillcolor(color)
    t.begin_fill()
    for _ in range(2):
        t.forward(ancho)
        t.left(90)
        t.forward(alto)
        t.left(90)
    t.end_fill()

def dibujar_triangulo(t, base, altura, color):
    t.fillcolor(color)
    t.begin_fill()
    t.forward(base)
    t.left(135)
    t.forward(altura)
    t.left(90)
    t.forward(altura)
    t.left(135)
    t.end_fill()

def dibujar_techo_con_lineas(x, y, base, altura, color="mistyrose"):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, base, altura, color)

    # Líneas naranjas simulando tejas
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pencolor("orange")
    for i in range(6):
        t.forward(base / 6)
        t.pendown()
        t.left(135)
        t.forward(altura)
        t.backward(altura)
        t.right(135)
        t.penup()
    t.pencolor("black")

def dibujar_maceta_con_flor(x, y, flor_color):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.color("saddlebrown")
    t.pendown()
    dibujar_rectangulo(t, 10, 5, "saddlebrown")
    t.penup()
    t.goto(x + 5, y + 5)
    t.setheading(0)
    t.color(flor_color)
    t.pendown()
    for _ in range(6):
        t.begin_fill()
        t.circle(3)
        t.end_fill()
        t.left(60)

def dibujar_sol(x, y):
    t.penup()
    t.goto(x, y)
    t.color("gold")
    t.begin_fill()
    t.circle(20)
    t.end_fill()
    for angle in range(0, 360, 30):
        t.setheading(angle)
        t.forward(30)
        t.backward(30)

#  Vistas
def vista_frontal(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 100, "mistyrose")
    dibujar_techo_con_lineas(x, y + 100, 100, 70)

    dibujar_rectangulo_desde(x + 35, y, 30, 50, "peru")

    t.penup()
    t.goto(x + 20, y + 85)
    t.color("black")
    t.write("Floristería Piña", font=("Arial", 8, "bold"))

    posiciones = [x + 5, x + 20, x + 75, x + 85]
    colores = ["red", "blue", "purple", "orange"]
    for px, color in zip(posiciones, colores):
        dibujar_maceta_con_flor(px, y + 10, color)

def vista_trasera(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 100, "mistyrose")
    dibujar_techo_con_lineas(x, y + 100, 100, 70)

    for px in range(x + 10, x + 100, 15):
        dibujar_maceta_con_flor(px, y + 10, "purple")

    # Sol centrado más abajo en el rectángulo
    dibujar_sol(x + 50, y + 50)

def vista_lateral(x, y, colores, con_ventana=False):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 60, 100, "mistyrose")
    dibujar_techo_con_lineas(x, y + 100, 60, 42)

    for i, color in enumerate(colores):
        dibujar_maceta_con_flor(x + 5 + i * 15, y + 10, color)

    if con_ventana:
        dibujar_rectangulo_desde(x + 20, y + 60, 20, 20, "white")

def vista_superior(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 80, "mistyrose")

    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pencolor("orange")
    for _ in range(10):
        t.forward(10)
        t.pendown()
        t.left(90)
        t.forward(80)
        t.backward(80)
        t.right(90)
        t.penup()
    t.pencolor("black")

# Auxiliar
def dibujar_rectangulo_desde(x, y, ancho, alto, color):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, ancho, alto, color)

# Composición final
vista_frontal(-250, 100)
vista_trasera(-80, 100)
vista_lateral(100, 100, ["blue", "magenta", "yellow", "deeppink"], con_ventana=True)
vista_lateral(220, 100, ["purple", "orange", "limegreen", "hotpink"])
vista_superior(-50, -150)

t.hideturtle()
screen.mainloop()
```
