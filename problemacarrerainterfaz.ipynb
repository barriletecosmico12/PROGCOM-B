```python
import tkinter as tk
from tkinter import messagebox
import numpy as np

# Configuramos el backend de Matplotlib para evitar errores
import matplotlib
matplotlib.use("TkAgg")
import matplotlib.pyplot as plt

def calcular_forward():
    try:
        precio = float(entry_precio.get())
        interes = float(entry_interes.get())
        tiempo = float(entry_tiempo.get())
        precioesperado = float(entry_precioesperado.get())

        # Forward discreto
        F_discreto = precio * (1 + interes) ** tiempo

        # Forward continuo
        F_continuo = precio * np.exp(interes * tiempo)

        # Recomendación
        if precioesperado > F_discreto:
            recomendacion = "Conviene comprar la acción hoy."
        elif precioesperado < F_discreto:
            recomendacion = "Conviene negociar con el Forward."
        else:
            recomendacion = "Ambas opciones son buenas."

        resultado = (
            f"Precio Forward (discreto): {F_discreto:.2f}\n"
            f"Precio Forward (continuo): {F_continuo:.2f}\n"
            f"Recomendación: {recomendacion}"
        )

        messagebox.showinfo("Resultado", resultado)

        # Gráfico
        tiempos = np.linspace(0, tiempo, 50)
        forwards = precio * (1 + interes) ** tiempos
        forwards_cont = precio * np.exp(interes * tiempos)

        plt.figure(figsize=(7, 5))
        plt.plot(tiempos, forwards, label="Forward discreto", color="blue")
        plt.plot(tiempos, forwards_cont, label="Forward continuo", color="red", linestyle="--")
        plt.axhline(precioesperado, color="green", linestyle=":", label="Precio esperado")
        plt.title("Evolución del Forward")
        plt.xlabel("Tiempo (años)")
        plt.ylabel("Precio Forward")
        plt.legend()
        plt.grid(True)
        plt.show()

    except ValueError:
        messagebox.showerror("Error", "Por favor ingrese valores numéricos válidos.")

# Interfaz gráfica
root = tk.Tk()
root.title("Calculadora de Precio Forward")

tk.Label(root, text="Precio spot del activo:").grid(row=0, column=0, padx=5, pady=5, sticky="e")
entry_precio = tk.Entry(root)
entry_precio.grid(row=0, column=1, padx=5, pady=5)

tk.Label(root, text="Tasa de interés anual (ej. 0.1 para 10%):").grid(row=1, column=0, padx=5, pady=5, sticky="e")
entry_interes = tk.Entry(root)
entry_interes.grid(row=1, column=1, padx=5, pady=5)

tk.Label(root, text="Tiempo al vencimiento (años):").grid(row=2, column=0, padx=5, pady=5, sticky="e")
entry_tiempo = tk.Entry(root)
entry_tiempo.grid(row=2, column=1, padx=5, pady=5)

tk.Label(root, text="Precio esperado al vencimiento:").grid(row=3, column=0, padx=5, pady=5, sticky="e")
entry_precioesperado = tk.Entry(root)
entry_precioesperado.grid(row=3, column=1, padx=5, pady=5)

tk.Button(root, text="Calcular Forward", command=calcular_forward).grid(row=4, column=0, columnspan=2, pady=10)

root.mainloop()
```
